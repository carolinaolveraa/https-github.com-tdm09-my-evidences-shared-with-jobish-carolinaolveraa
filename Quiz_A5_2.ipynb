```python
 import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()
df = pd.read_csv(list(uploaded.keys())[0])

hombres = df[df['sexo'] == 1]
mujeres = df[df['sexo'] == 2]
```

```python
plt.boxplot(df['peso'].dropna())
plt.title('Boxplot de peso')
plt.show()

plt.plot(df['peso'].dropna())
plt.title('Line Plot de peso')
plt.xlabel('Índice')
plt.ylabel('Peso')
plt.show()

plt.hist(df['peso'].dropna(), bins=20)
plt.title('Histograma de peso')
plt.xlabel('Peso')
plt.ylabel('Frecuencia')
plt.show()

plt.boxplot(df['cintura'].dropna())
plt.title('Boxplot de cintura')
plt.show()

plt.plot(df['cintura'].dropna())
plt.title('Line Plot de cintura')
plt.xlabel('Índice')
plt.ylabel('Cintura')
plt.show()

plt.hist(df['cintura'].dropna(), bins=20)
plt.title('Histograma de cintura')
plt.xlabel('Cintura')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.hist(hombres['peso'], bins=20, alpha=0.5, label='Hombres')
plt.hist(mujeres['peso'], bins=20, alpha=0.5, label='Mujeres')
plt.title('Histograma de peso por sexo')
plt.legend()
plt.show()

plt.hist(hombres['cintura'], bins=20, alpha=0.5, label='Hombres')
plt.hist(mujeres['cintura'], bins=20, alpha=0.5, label='Mujeres')
plt.title('Histograma de cintura por sexo')
plt.legend()
plt.show()
```

```python
plt.scatter(df['peso'], df['cintura'])
plt.title('Diagrama de dispersión: Peso vs Cintura')
plt.xlabel('Peso')
plt.ylabel('Cintura')
plt.show()
```

```python
plt.scatter(hombres['peso'], hombres['cintura'], alpha=0.5, label='Hombres')
plt.scatter(mujeres['peso'], mujeres['cintura'], alpha=0.5, label='Mujeres')
plt.title('Dispersión Peso vs Cintura por sexo')
plt.xlabel('Peso')
plt.ylabel('Cintura')
plt.legend()
plt.show()
```

# **Conclusión**
El análisis muestra que las variables peso y cintura siguen patrones similares: ambas son asimétricas, presentan valores atípicos, y tienen una correlación positiva fuerte.
