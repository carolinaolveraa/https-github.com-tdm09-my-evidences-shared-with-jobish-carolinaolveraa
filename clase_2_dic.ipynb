**Regression**

```python
path="https://raw.githubusercontent.com/tec03/Datasets/refs/heads/main/datasets/reg_data.csv"
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
df = pd.read_csv(path)
df.head(2)
```

1. Clean data and prepare dummy variables

```python
df2 = df.copy()
df2 = df2.dropna()
df2 = pd.get_dummies(df2, columns=["Gender","Company","Place","Education"], drop_first=True)
df2.head()
```

2. Creating and training the regression model

```python
X = df2.drop(["Salary"], axis=1)
y = df2["Salary"]

model = LinearRegression()
model.fit(X, y)

model.intercept_, model.coef_

```

```python
df2.columns
```

What is the salary of a person with age 33?

Salary=b0​+b1​(Age)

```python
x = np.zeros((1, X.shape[1]))
x[0, X.columns.get_loc("Age")] = 33
print("The salary of a person with age 33 is:", model.predict(x)[0])
```
