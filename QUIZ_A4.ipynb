```python
import pandas as pd
from google.colab import files

uploaded = files.upload()  # Upload antropometria-dataset-2.csv and adult.csv

df = pd.read_csv('antropometria-dataset-2.csv')
adult = pd.read_csv('adult.csv')

df['desc_ent'] = (df['desc_ent']
    .astype(str)
    .str.normalize('NFKD')
    .str.encode('ascii', errors='ignore')
    .str.decode('utf-8')
    .str.strip()
    .str.upper())

adult = adult.applymap(lambda x: x.strip().upper() if isinstance(x, str) else x)

pivot_all = df.pivot_table(index='desc_ent', values='peso', aggfunc='mean')
pivot_s1 = df[df['sexo'] == 1].pivot_table(index='desc_ent', values='peso', aggfunc='mean')
pivot_s2 = df[df['sexo'] == 2].pivot_table(index='desc_ent', values='peso', aggfunc='mean')

pivot = pivot_all.join(pivot_s1, rsuffix='_1').join(pivot_s2, rsuffix='_2')
pivot.columns = ['All', 'Sexo1', 'Sexo2']

print("=== Average Weight by State and Sex ===")
print(pivot)

sinaloa_avg = pivot.loc['SINALOA', 'All']
print("\n✅ Average weight of people from SINALOA:", sinaloa_avg)

jalisco_median_s1 = df[(df['desc_ent'] == 'JALISCO') & (df['sexo'] == 1)]['peso'].median()
print("\n✅ Median weight (peso) of sex1 of JALISCO:", jalisco_median_s1)

lowest_s1_state = pivot['Sexo1'].idxmin()
lowest_s1_value = pivot['Sexo1'].min()
print("\n✅ State with lowest average peso (weight) of sex1:", lowest_s1_state, lowest_s1_value)

mediana_edad_10th = adult[
    (adult['sex'] == 'FEMALE') &
    (adult['marital-status'] == 'DIVORCED') &
    (adult['education'] == '10TH')
]['age'].median()
print("\n✅ Median age of Female Divorced with education 10th:", mediana_edad_10th)

promedio_horas_bach = adult[
    (adult['marital-status'] == 'NEVER-MARRIED') &
    (adult['education'] == 'BACHELORS')
]['hours-per-week'].mean()
print("\n✅ Average hours-per-week of Never-married with Bachelors:", promedio_horas_bach)

yucatan_median_s2 = df[(df['desc_ent'] == 'YUCATAN') & (df['sexo'] == 2)]['peso'].median()
print("\n✅ Median weight (peso) of sex2 in YUCATAN:", yucatan_median_s2)

min_avg_s2 = pivot['Sexo2'].min()
print("\n✅ Minimum average weight of sex2 across states:", min_avg_s2)

promedio_horas_hs = adult[
    (adult['marital-status'] == 'DIVORCED') &
    (adult['education'] == 'HS-GRAD')
]['hours-per-week'].mean()
print("\n✅ Average hours-per-week of Divorced with HS-grad:", promedio_horas_hs)

mediana_edad_12th = adult[
    (adult['sex'] == 'FEMALE') &
    (adult['marital-status'] == 'DIVORCED') &
    (adult['education'] == '12TH')
]['age'].median()
print("\n✅ Median age of Female Divorced with education 12th:", mediana_edad_12th)

```
