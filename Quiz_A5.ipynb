```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

uploaded = files.upload()
df = pd.read_csv("antropometria-dataset-1-1.csv")

plt.figure(figsize=(8,4))
sns.boxplot(data=df[['peso', 'cintura']])
plt.title('Boxplot de Peso y Cintura')
plt.show()

df[['peso', 'cintura']].hist(bins=30, figsize=(10,4), color='skyblue', edgecolor='black')
plt.suptitle('Histogramas de Peso y Cintura')
plt.show()

plt.figure(figsize=(8,4))
sns.lineplot(x='edad', y='peso', data=df.sort_values('edad'), color='green')
plt.title('Evolucin del peso segn edad')
plt.xlabel('Edad')
plt.ylabel('Peso')
plt.show()

plt.figure(figsize=(8,4))
sns.boxplot(x='sexo', y='peso', data=df)
plt.title('Distribucin del Peso por Sexo')
plt.xlabel('Sexo (1 = Hombre, 2 = Mujer)')
plt.show()

plt.figure(figsize=(8,4))
sns.boxplot(x='sexo', y='cintura', data=df)
plt.title('Distribucin de la Cintura por Sexo')
plt.xlabel('Sexo (1 = Hombre, 2 = Mujer)')
plt.show()

plt.figure(figsize=(8,4))
sns.histplot(data=df, x='peso', hue='sexo', bins=30, kde=True)
plt.title('Histograma del Peso por Sexo')
plt.show()

plt.figure(figsize=(8,4))
sns.histplot(data=df, x='cintura', hue='sexo', bins=30, kde=True)
plt.title('Histograma de la Cintura por Sexo')
plt.show()

plt.figure(figsize=(6,5))
sns.scatterplot(data=df, x='cintura', y='peso', hue='sexo', alpha=0.6)
plt.title('Relacin entre Peso y Cintura por Sexo')
plt.xlabel('Cintura')
plt.ylabel('Peso')
plt.show()
```
