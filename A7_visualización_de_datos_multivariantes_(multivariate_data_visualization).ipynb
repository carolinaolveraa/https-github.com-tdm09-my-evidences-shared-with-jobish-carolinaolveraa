```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt

uploaded = files.upload()
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

df.head()

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.hist(df['sistol'], bins=20)
plt.title('Histograma de Sistol')
plt.subplot(1,2,2)
plt.hist(df['diastol'], bins=20)
plt.title('Histograma de Diastol')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.hist(df['peso'], bins=20)
plt.title('Histograma de Peso')
plt.subplot(1,2,2)
plt.hist(df['cintura'], bins=20)
plt.title('Histograma de Cintura')
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df['peso'], df['cintura'])
plt.xlabel('Peso (kg)')
plt.ylabel('Cintura (cm)')
plt.title('Relacin entre Peso y Cintura')
plt.show()

hombres = df[df['sexo'] == 1]
mujeres = df[df['sexo'] == 2]

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.hist(hombres['peso'], bins=20)
plt.title('Peso (Hombres)')
plt.subplot(1,2,2)
plt.hist(hombres['cintura'], bins=20)
plt.title('Cintura (Hombres)')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.hist(mujeres['peso'], bins=20)
plt.title('Peso (Mujeres)')
plt.subplot(1,2,2)
plt.hist(mujeres['cintura'], bins=20)
plt.title('Cintura (Mujeres)')
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(hombres['peso'], hombres['cintura'])
plt.xlabel('Peso (Hombres)')
plt.ylabel('Cintura')
plt.title('Peso vs Cintura (Hombres)')
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(mujeres['peso'], mujeres['cintura'])
plt.xlabel('Peso (Mujeres)')
plt.ylabel('Cintura')
plt.title('Peso vs Cintura (Mujeres)')
plt.show()
```

**1. Cmo son las distribuciones?**

Sistol y diastol: distribuciones asimtricas a la derecha, con una mayora de valores normales y una cola hacia lecturas altas.

Peso y cintura: tambin asimtricas a la derecha, con presencia de personas con valores elevados que generan ese sesgo.

**2. Existen valores atpicos en los histogramas?**

S. En todas las variables aparecen valores alejados del grupo principal, especialmente: pesos altos, cinturas grandes, presiones sistlica y diastlica elevadas. Estos valores producen colas largas en los histogramas.

**3. Cmo es el comportamiento del peso con respecto a la cintura?**

Existe una relacin positiva clara: conforme aumenta el peso, aumenta la cintura. El diagrama de dispersin muestra una tendencia ascendente que indica correlacin directa entre ambas variables.

**4. Hay valores atpicos en el scatter pesocintura?**

S. Se observan puntos separados del patrn general: personas con cintura elevada pero peso moderado, personas con peso alto pero cintura y proporcionalmente menor

Estos puntos se interpretan como outliers.

**5. Anlisis al separar por sexo**

- Hombres:

Valores ms altos de peso y cintura.

Mayor dispersin.

Distribuciones sesgadas a la derecha.

- Mujeres:

Rangos menores y ms concentrados.

Tambin sesgo a la derecha, pero con menos variabilidad.

Conclusin conjunta:


En ambos sexos, peso y cintura mantienen una relacin positiva.

La correlacin es ms marcada en hombres por la mayor amplitud de valores.

En mujeres la relacin es ms estable y uniforme, dentro de rangos ms bajos.
